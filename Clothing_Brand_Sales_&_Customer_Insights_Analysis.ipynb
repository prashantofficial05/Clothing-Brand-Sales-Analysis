```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/articles_hm.csv')
data
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data.describe()
```

Top-Selling Product Categories

```python

top_products = data['product_type_name'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_products.values, y=top_products.index, palette='coolwarm')
plt.xlabel("Number of Articles")
plt.ylabel("Product Type")
plt.title("Top 10 Selling Product Types")
plt.show()
```

Most Popular Colors

```python

top_colors = data['colour_group_name'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_colors.values, y=top_colors.index, palette='viridis')
plt.xlabel("Number of Articles")
plt.ylabel("Color")
plt.title("Top 10 Popular Colors")
plt.show()
```

Analyzing Garment Groups

```python

top_garments = data['garment_group_name'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_garments.values, y=top_garments.index, palette='magma')
plt.xlabel("Number of Articles")
plt.ylabel("Garment Group")
plt.title("Top Garment Groups")
plt.show()
```

Graphical Appearance Trends

```python

top_graphics = data['graphical_appearance_name'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_graphics.values, y=top_graphics.index, palette='cividis')
plt.xlabel("Number of Articles")
plt.ylabel("Graphical Appearance")
plt.title("Most Common Graphical Appearances")
plt.show()
```
